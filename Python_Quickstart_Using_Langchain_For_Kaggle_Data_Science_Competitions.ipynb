# 8. Here's where the magic is - Langchain Agents
## Let's install Langchain Agents that are specially designed to handle csv database files

```python
from langchain.agents import create_csv_agent
from langchain.llms import OpenAI
```

# 9. Now that we have the Langchain Agent installed, let's give it the Disney Titles data.

```python
agent = create_csv_agent(OpenAI(temperature=0), 
                         '/content/titles.csv', 
                         verbose=True)
```

# 10. Let's run the Langchain Agent

```python
agent
```

# 11. Here's where the magic continues to get even more interesting.
 Langchain has pre-created agents that will handle database data
## Read through that text below and see how Langchain will create a Data Scientist that works FOR YOU.

```python
agent.agent.llm_chain.prompt.template
```
